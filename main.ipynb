### Função para verificar os casos nulos

```python
df_cln = df.copy()
```

```python
def check_missing(df_param):
    res_missing = df_param.isna().sum()
    res_missing = (res_missing/len(df_param))*100
    return res_missing
```

```python
check_missing(df_cln).sort_values(ascending=False)[:20]
```

```python
list_nulos = [
    'CPF_AUT', 'INFEHOSP', 'GESTOR_DT', 'NUM_PROC', 'DIAGSEC9', 'DIAGSEC8',
    'CID_NOTIF', 'FAEC_TP', 'DIAGSEC7', 'DIAGSEC6', 'DIAGSEC5', 'SIS_JUST',
    'DIAGSEC4', 'DIAGSEC3', 'DIAGSEC2', 'DIAGSEC1'
]
for column in list_nulos: 
    df_cln[column] = df_cln[column].fillna("no_info")
```

```python
list_nulos = ['CNPJ_MANT', 'CGC_HOSP']
for column in list_nulos: 
    df_cln[column] = df_cln[column].fillna("0")
```

```python
check_missing(df_cln).sort_values(ascending=False)[:20]
```

### Eliminando as colunas constantes

### Eliminando as colunas extremamente desbalanceadas

```python
list_1 = []
limit = 0.98
for col in df_cln.columns:
    perc = df_cln[col].value_counts(normalize=True).values[0]
    if perc > limit:
        list_1.append(col)
        print(col, perc)
```

```python
print("Antes: ", len(df_cln.columns))
df_cln = df_cln.drop(list_1, axis=1)
print("Depois: ", len(df_cln.columns))
```

### Chamada do "contexto"

```python
import great_expectations as gx

context = gx.get_context()
print(context)
```

```python
context.list_datasources()
```

```python
my_connection_string = (
    #"postgresql+psycopg2://<username>:<password>@<host>:<port>/<database>"
    "postgresql+psycopg2://root:root@localhost:5432/sus-dbt"
)
```

```python
datasource = context.sources.add_postgres(name="ge_datasource", connection_string=my_connection_string)
```

```python
context.list_datasources()
```

### Amostra "Bacht"

```python
asset_name = "silver"
asset_table_nome = "sus_silver"

table_asset = datasource.add_table_asset(name=asset_name, table_name=asset_table_nome)
```

```python
context.list_datasources()
```

```python
my_datasource = context.get_datasource("ge_datasource")
my_table_asset = my_datasource.get_asset(asset_name="silver")
batch_request = my_table_asset.build_batch_request()
```

```python
context.add_or_update_expectation_suite("suite_silver")
```

### Cria um avaliador

```python
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = "suite_silver",
)
```

```python
checkpoint = context.add_or_update_checkpoint(
    name="checkpoint_silver",
    validator=validator
)
```

```python
checkpoint_result = checkpoint.run()
```

```python
validator.head(10)
```

```python
validator.expect_column_values_to_be_of_type(column="CGC_HOSP", type_='INTEGER')
validator.expect_column_values_to_be_of_type(column="CNPJ_MANT", type_='INTEGER')
validator.expect_column_values_to_not_be_null(column="CEP")
validator.expect_column_values_to_be_between(column="VAL_SH", min_value=1, max_value=1000)
validator.expect_column_values_to_be_between(column="IDADE", min_value=20, max_value=40)
```

```python
validator.save_expectation_suite(discard_failed_expectations=False)
```

```python
checkpoint = context.add_or_update_checkpoint(name="checkpoint_silver", validator=validator)
```

```python
checkpoint_result = checkpoint.run()
```
